```python
hevlay_ats = scsd.import_pdb(
    "C:\\Users\\ckingsbury\\scsd\\example\\notebooks\\data\\HEVLAY.pdb"
)
model_ag_ats = md.ats.copy()
model_ag_ats[md.ats == "AU"] = "AG"
model_ag = scsd.scsd_model(
    "hakqao_ag", model_ag_ats, "D2d",
    maxdist=2.1,
    mondrian_limits=[0.0, 0.0],
)

hevlay_scsd = scsd.scsd_matrix(hevlay_ats, model=model_ag, ptgr="D2d")
hevlay_scsd.calc_scsd(bhopping=True)
```
